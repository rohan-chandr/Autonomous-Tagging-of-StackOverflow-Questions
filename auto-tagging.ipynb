```python
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 

from util import encoder, question_tagger, normalizer as norm

inputs = 50000
train_ratio = int(inputs / 0.9)
test_ratio = int(inputs / 0.9 * 0.1)
classes = 50
lr = 0.001
epochs = 5000
steps = 100
hidden_layer_sizes = [12, 14, 26, 46]
hidden_layers = len(hidden_layer_sizes)



def model():
    # define the layers and parameters of the neural net
    weights = [random(input_size, hidden_layer_sizes[0])]
    biases = [zeros(hidden_layer_sizes[0])]
    layers = [tf.nn.relu(tf.add(tf.matmul(ph_in, weights[0]), biases[0]))]
    test_layers = [tf.nn.relu(tf.add(tf.matmul(ph_test_in, weights[0]), biases[0]))]


    for i in range(1, len(hidden_layer_sizes)):
        weights.append(random(hidden_layer_sizes[i - 1], hidden_layer_sizes[i]))
        biases.append(zeros(hidden_layer_sizes[i]))
        layers.append(tf.nn.relu(tf.add(tf.matmul(layers[i - 1], weights[i]), biases[i])))
        test_layers.append(tf.nn.relu(tf.add(tf.matmul(test_layers[i - 1], weights[i]), biases[i])))

    weights.append(random(hidden_layer_sizes[hidden_layers - 1], classes))
    biases.append(zeros(classes))
    layers.append(tf.nn.softmax(tf.add(tf.matmul(layers[hidden_layers - 1], weights[hidden_layers]), biases[hidden_layers])))
    test_layers.append(tf.nn.softmax(tf.add(tf.matmul(test_layers[hidden_layers - 1], weights[hidden_layers]), biases[hidden_layers])))

    return layers[hidden_layers], test_layers[hidden_layers]



if __name__ == "__main__":

    # import data, one hot encode the output classes 
    csv_input, csv_output = question_tagger._import('data/tagged_questions.csv', classes)
    for i in range(0, len(csv_output)):
        csv_output[i] = encoder.one_hot(int(csv_output[i]), classes)
    input_size = len(csv_input[0])

    # split data into training and test setTh
    train_in, train_out = csv_input[:train_ratio], csv_output[:train_ratio]
    test_in, test_out = csv_input[train_ratio:train_ratio + test_ratio], csv_output[train_ratio:train_ratio + test_ratio]
    train_in, test_in = norm.standard(train_in, test_in)

    # set placeholders in tensorflow
    ph_in = tf.placeholder(tf.float32, shape=[train_ratio, input_size])
    ph_out = tf.placeholder(tf.float32, shape=[train_ratio, classes])
    ph_test_in = tf.placeholder(tf.float32, shape=[test_ratio, input_size])
    ph_test_out = tf.placeholder(tf.float32, shape=[test_ratio, classes])

    net, test_net = model()

    # MSE calculation 
    LOSS = tf.nn.l2_loss(net - ph_out)
    accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(net, 1), tf.argmax(ph_out, 1)), tf.float32))

    # initialize our optimizer 
    optimizer = tf.train.AdamOptimizer(lr).minimize(LOSS)

    sess = tf.Session()
    init = tf.global_variables_initializer()
    sess.run(init)

    losses = []
    epoch = 0

    # train, plot the results along with test results 
    acc_train = sess.run(accuracy, feed_dict={ph_in: train_in, ph_out: train_out})
    while epoch < epochs:
        train_dict = {ph_in: train_in, ph_out: train_out}
        test_dict = {ph_test_in: test_in, ph_test_out: test_out}
        sess.run(optimizer, feed_dict=train_dict)
        epoch += 1

        if epoch % steps == 0:
            acc_train = sess.run(accuracy, feed_dict=train_dict)
            acc_test = sess.run(tf.reduce_mean(
                tf.cast(tf.equal(tf.argmax(test_net, 1), tf.argmax(ph_test_out, 1)), tf.float32)), feed_dict=test_dict)
            actual_loss = sess.run(LOSS, feed_dict=train_dict)
            losses.append(actual_loss)
            plt.plot(losses, 'b')
            plt.ylabel('Loss ')
            plt.draw()
            plt.pause(0.001)
            print('Epoch: ' + str(epoch) + ', Loss: ' + str(actual_loss) + ', Train Accuracy: ' +
                str(acc_train) + ', Test Accuracy: ' + str(acc_test))


```
